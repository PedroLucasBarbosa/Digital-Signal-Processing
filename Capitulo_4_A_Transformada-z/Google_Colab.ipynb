####Comentários

*   A variável complexa _Z_ é chamada de **Frequência complexa**, dada por **z = |z|ejω**, em que |_z_| é a magnitude e o ω é a frequência real.

* Por ser definifido por magnitude, a forma do ROC(região de convergência) é um anel.

* O ROC(região de convergência) é _null space_ e a transformada-_z_ não existe, se Rx+ < Rx−.



####**Propriedades da Região de convergência**

Observando a **ROC**, é possível notar que:

* Quando a convegência tem a condiçao de magnitude _|z|_, o ROC será sempre delimitado por um círculo.

* Existem dois casos especiais, a sequência right-sided e o left-sided.

**>**A sequência right-sided, é definida como uma sequência _X(n)_ que é zero para _n < n0_. O ROC para esse caso é sempre fora de um círculo de raio _Rx-_. **Se _n0 ≥ 0_, também pode ser chamado de sequência casual**.

**>**A sequêcia left-sided, é definida como uma sequência _X(n)_ que pelo contrario do right-sided, o zero para _n > n0_. O ROC para esse caso é sempre dentro do círculo do raio _Rx+_. **Se _n0 ≤ 0_, também é chamado de sequência anti-casual.**

* Há também a sequência two-sided. O ROC para esse tipo é sempre um anel, _Rx− < |z| < Rx+_

* Em PDS, assume-se sinais casuais, pois quase todo dado digital pe adquirido em tempo real. Logo, o único ROC que realmente vai importar pra esse fim será a sequência right-sided.

````python
import numpy as np
import matplotlib.pyplot as plt

# Função equivalente a conv_m no MATLAB
def conv_m(x, nx, h, nh):
    # Calculando os índices para a convolução
    nyb = nx[0] + nh[0]  # Índice inicial
    nye = nx[-1] + nh[-1]  # Índice final
    ny = np.arange(nyb, nye + 1)  # Vetor de índices

    # Calculando a convolução
    y = np.convolve(x, h, mode='full')  # Convolução dos sinais

    return y, ny  # Retorna o resultado da convolução e os índices

# Definindo os sinais x1 e x2 com seus respectivos índices
x1 = [1, 2, 3]
n1 = np.array([-1, 0, 1])
x2 = [2, 4, 3, 5]
n2 = np.array([-2, -1, 0, 1])

# Calculando a convolução e os índices
x3, n3 = conv_m(x1, n1, x2, n2)

# Exibindo os resultados
print("x3:", x3)
print("n3:", n3)

# Plotando os sinais e a convolução
plt.figure(figsize=(10, 8))

# Gráfico do sinal x1
plt.subplot(3, 1, 1)
plt.stem(n1, x1, basefmt=" ")
plt.title('Sinal x1')
plt.xlabel('n1')
plt.ylabel('x1')
plt.grid()

# Gráfico do sinal x2
plt.subplot(3, 1, 2)
plt.stem(n2, x2, basefmt=" ")
plt.title('Sinal x2')
plt.xlabel('n2')
plt.ylabel('x2')
plt.grid()

# Gráfico da convolução x3
plt.subplot(3, 1, 3)
plt.stem(n3, x3, basefmt=" ")
plt.title('Resultado da Convolução x3')
plt.xlabel('n3')
plt.ylabel('x3')
plt.grid()

# Ajustando o layout
plt.tight_layout()
plt.show()

````

````python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

# Função equivalente a impseq
def impseq(n0, n1, n2):
    if n0 < n1 or n0 > n2 or n1 > n2:
        raise ValueError('Arguments must satisfy n1 <= n0 <= n2')

    n = np.arange(n1, n2+1)
    x = (n - n0) == 0  # Sequência impulso unitário
    return x, n

# Função stepseq, para criar uma sequência de passo unitário (semelhante ao stepseq de MATLAB)
def stepseq(n0, n1, n2):
    n = np.arange(n1, n2+1)
    x = n >= n0
    return x, n

# Parâmetros
b = [0, 0, 0, 0.25, -0.5, 0.0625]
a = [1, -1, 0.75, -0.25, 0.0625]

# Gerar a sequência impulso unitário
delta, n = impseq(0, 0, 7)

# Aplicando o filtro
x_filtered = signal.lfilter(b, a, delta)

# Gerar a sequência original usando a expressão dada
x_original = ((n - 2) * (1/2)**(n - 2) * np.cos(np.pi * (n - 2) / 3)) * stepseq(2, 0, 7)[0]

# Plotando os resultados
plt.figure(figsize=(10, 6))

# Sequência filtrada
plt.subplot(2, 1, 1)
plt.stem(n, x_filtered)
plt.title('Sequência Filtrada')
plt.xlabel('n')
plt.ylabel('x_filtered[n]')

# Sequência original
plt.subplot(2, 1, 2)
plt.stem(n, x_original)
plt.title('Sequência Original')
plt.xlabel('n')
plt.ylabel('x_original[n]')

plt.tight_layout()
plt.show()


````

####SIMULAÇÃO 4.8



````python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

# Coeficientes de b e a
b = np.array([0, 1])
a = np.array([3, -4, 1])

# Decomposição em frações parciais
R, p, C = signal.residuez(b, a)

# Exibindo os resultados
print("R (Resíduos):", R)
print("p (Polos):", p)
print("C (Coeficientes):", C)

# Reconstrução do numerador
numerador = np.zeros(len(a) - 1)
for i in range(len(R)):
    numerador += R[i] * np.poly([p[i]])  # Poly com lista contendo o polo

# Reconstrução do denominador
denominador = np.poly(p)

# Exibindo os coeficientes reconstruídos
print("\nNumerador (Reconstruído):", numerador)
print("Denominador (Reconstruído):", denominador)

# Plotando a resposta em frequência do sistema reconstruído
w, H = signal.freqz(numerador, denominador, 512)  # 512 pontos para a resposta de frequência
magH = np.abs(H)
phaH = np.angle(H) / np.pi  # Fase em unidades de pi

# Figura para a resposta de magnitude e fase
plt.figure(figsize=(10, 6))

# Subplot para magnitude
plt.subplot(2, 1, 1)
plt.plot(w / np.pi, magH)
plt.title('Resposta de Magnitude')
plt.xlabel('Frequência (unidades de π)')
plt.ylabel('Magnitude')
plt.grid()

# Subplot para fase
plt.subplot(2, 1, 2)
plt.plot(w / np.pi, phaH)
plt.title('Resposta de Fase')
plt.xlabel('Frequência (unidades de π)')
plt.ylabel('Fase (unidades de π)')
plt.grid()

plt.tight_layout()
plt.show()

# Plotando polos e resíduos
plt.figure(figsize=(6, 6))
plt.scatter(np.real(p), np.imag(p), color='red', label='Polos (p)')
plt.scatter(np.real(R), np.imag(R), color='blue', label='Resíduos (R)')
plt.title('Plano de Polos e Resíduos')
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginária')
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid()
plt.legend()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import residuez, lfilter

# Definindo o numerador e denominador
b = np.array([1, 0.4 * np.sqrt(2)])
a = np.array([1, -0.8 * np.sqrt(2), 0.64])

# Decomposição em frações parciais
R, p, C = residuez(b, a)

# Exibindo os resíduos, polos e coeficientes
print("Resíduos (R):", R)
print("Polos (p):", p)
print("Coeficientes (C):", C)

# Magnitude e ângulo dos polos
Mp = np.abs(p)  # Magnitude dos polos
Ap = np.angle(p) / np.pi  # Ângulo dos polos em unidades de pi

print("\nMagnitude dos polos (Mp):", Mp)
print("Ângulo dos polos (Ap):", Ap)

# Gerando sequência delta (impulso unitário) de 0 a 6
n = np.arange(0, 7)
delta = (n == 0).astype(int)

# Filtrando a sequência delta
x_filtered = lfilter(b, a, delta)

# Exibindo a sequência filtrada
print("\nSequência filtrada (x):", x_filtered)

# Resposta manual do sistema (frações parciais)
x_response = (0.8)**n * (np.cos(np.pi * n / 4) + 2 * np.sin(np.pi * n / 4))

# Exibindo a resposta manual do sistema
print("\nResposta manual (x_response):", x_response)

# Plotando os resultados
plt.figure(figsize=(10, 6))

# Plot da sequência filtrada
plt.subplot(2, 1, 1)
plt.stem(n, x_filtered)
plt.title('Sequência Filtrada')
plt.xlabel('n')
plt.ylabel('x[n]')

# Plot da resposta manual
plt.subplot(2, 1, 2)
plt.stem(n, x_response)
plt.title('Resposta Manual')
plt.xlabel('n')
plt.ylabel('x[n]')

plt.tight_layout()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import residuez, lfilter

# Coeficientes e condições iniciais
b = np.array([1, 1, 1]) / 3
a = np.array([1, -0.95, 0.9025])
Y = np.array([-2, -3])
X = np.array([1, 1])
xic = np.array([1.4742, 2.1383])

bxplus = np.array([1, -0.5])
axplus = np.array([1, -1, 1])

# Transformação de X(z) para Y+(z)
ayplus = np.convolve(a, axplus)
byplus = np.convolve(b, bxplus) + np.convolve(xic, axplus)

# Decomposição em frações parciais
R, p, C = residuez(byplus, ayplus)

# Cálculo das magnitudes e ângulos dos polos
Mp = np.abs(p)
Ap = np.angle(p) / np.pi

# Definindo o sinal de entrada
n = np.arange(0, 8)
x = np.cos(np.pi * n / 3)

# Filtragem (corrigido para lidar com a condição inicial)
# Passando a condição inicial `xic` como uma tupla
y, _ = lfilter(b, a, x, zi=xic)

# Reconstrução usando resíduos e polos
A = np.real(2 * R[0])
B = np.imag(2 * R[0])
C = np.real(2 * R[2])
D = np.imag(2 * R[3])
y_reconstructed = A * np.cos(np.pi * n / 3) + B * np.sin(np.pi * n / 3) + (0.95) ** n * (C * np.cos(np.pi * n / 3) + D * np.sin(np.pi * n / 3))

# Plotando os resultados
plt.figure(figsize=(10, 8))

# Gráfico da sequência filtrada
plt.subplot(2, 1, 1)
plt.stem(n, y, basefmt=" ")
plt.title('Resposta Filtrada (y)')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.grid(True)

# Gráfico da resposta reconstruída
plt.subplot(2, 1, 2)
plt.stem(n, y_reconstructed, basefmt=" ")
plt.title('Resposta Reconstruída (y reconstruído)')
plt.xlabel('n')
plt.ylabel('y[n] Reconstruído')
plt.grid(True)

plt.tight_layout()
plt.show()
````
