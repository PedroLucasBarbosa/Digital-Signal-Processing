FILTROS PASSA-BAIXA ELÍPTICOS

Esses filtros exibem comportamento equiripple na banda passante, bem como na
faixa de parada. Eles são semelhantes em magnitude às características de resposta aos filtros equiripple FIR. Portanto, os filtros elípticos são filtros ideais, na medida em que eles alcançam a ordem mínima N para as especificações fornecidas (ou, alternativamente, alcance a banda de transição mais nítida para a ordem N). Esses filtros, por razões óbvias, são muito difíceis de analisar e, portanto, para projetar. Não é possível projetá-los usando ferramentas simples, e frequentemente é necessário fazer uso de programas ou tabelas para tal. A resposta de magnitude quadrada de filtros elípticos é dada por:

###SIMULAÇÃO 8.1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, freqresp, impulse, lti

# Parâmetros do filtro
Wp = 0.2 * np.pi  # Frequência de passagem
Ws = 0.3 * np.pi  # Frequência de rejeição
Rp = 7            # Ripple na banda de passagem (dB)
As = 16           # Atenuação na banda de rejeição (dB)

# Conversão de Ripple e Atenuação
Ripple = 10 ** (-Rp / 20)
Attn = 10 ** (-As / 20)

# Projeto do filtro Butterworth analógico
N, Wn = 3, Wp  # Ordem do filtro e frequência normalizada (definida fixamente como exemplo)
b, a = butter(N, Wn, analog=True)
print("*** Butterworth Filter Order = 3 ***")
print("Numerador (b):", b)
print("Denominador (a):", a)

# Cálculo da resposta em frequência
w = np.linspace(0, 0.5 * np.pi, 512)
_, H = freqresp((b, a), w=w)
magnitude = np.abs(H)
dB = 20 * np.log10(magnitude)
phase = np.angle(H)

# Cálculo da resposta ao impulso
system = lti(b, a)
t, h = impulse(system)

# Plot da resposta em frequência
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.plot(w / np.pi, dB, 'b')
plt.title('Resposta em Frequência do Filtro Butterworth')
plt.xlabel('Frequência Normalizada (x π rad/sample)')
plt.ylabel('Magnitude (dB)')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(w / np.pi, phase, 'r')
plt.title('Fase da Resposta em Frequência')
plt.xlabel('Frequência Normalizada (x π rad/sample)')
plt.ylabel('Fase (radianos)')
plt.grid()

# Plot da resposta ao impulso
plt.figure(figsize=(8, 6))
plt.stem(t, h, basefmt=" ")
plt.title('Resposta ao Impulso do Filtro Butterworth')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

# Parâmetros do filtro
Wp = 0.2 * np.pi  # Frequência de passagem (rad/s)
Ws = 0.3 * np.pi  # Frequência de rejeição (rad/s)
Rp = 7            # Ripple na banda de passagem (dB)
As = 16           # Atenuação na banda de rejeição (dB)

# Conversão de Ripple e Atenuação para fatores lineares
Ripple = 10 ** (-Rp / 20)
Attn = 10 ** (-As / 20)

# Cálculo da ordem do filtro e da frequência de corte
N, Wn = signal.buttord(Wp/np.pi, Ws/np.pi, Rp, As)

# Coeficientes do filtro Butterworth
b, a = signal.butter(N, Wn, btype='low')

# Cálculo da resposta em frequência
w, h = signal.freqs(b, a, worN=2000)

# Cálculo da resposta ao impulso
t, impulse_response = signal.impulse((b, a))

# Gráfico da resposta em frequência
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(w * np.pi, 20 * np.log10(abs(h)), 'b')
plt.title('Resposta em Frequência do Filtro IIR Butterworth')
plt.xlabel('Frequência (rad/s)')
plt.ylabel('Magnitude (dB)')
plt.grid()

# Gráfico da resposta ao impulso
plt.subplot(2, 1, 2)
plt.stem(t, impulse_response, basefmt=" ", linefmt="r")
plt.title('Resposta ao Impulso do Filtro IIR Butterworth')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid()

plt.tight_layout()
plt.show()
```
