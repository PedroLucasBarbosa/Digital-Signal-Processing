###SIMULAÃ‡ÃƒO 8.1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, freqresp, impulse, lti

# ParÃ¢metros do filtro
Wp = 0.2 * np.pi  # FrequÃªncia de passagem
Ws = 0.3 * np.pi  # FrequÃªncia de rejeiÃ§Ã£o
Rp = 7            # Ripple na banda de passagem (dB)
As = 16           # AtenuaÃ§Ã£o na banda de rejeiÃ§Ã£o (dB)

# ConversÃ£o de Ripple e AtenuaÃ§Ã£o
Ripple = 10 ** (-Rp / 20)
Attn = 10 ** (-As / 20)

# Projeto do filtro Butterworth analÃ³gico
N, Wn = 3, Wp  # Ordem do filtro e frequÃªncia normalizada (definida fixamente como exemplo)
b, a = butter(N, Wn, analog=True)
print("*** Butterworth Filter Order = 3 ***")
print("Numerador (b):", b)
print("Denominador (a):", a)

# CÃ¡lculo da resposta em frequÃªncia
w = np.linspace(0, 0.5 * np.pi, 512)
_, H = freqresp((b, a), w=w)
magnitude = np.abs(H)
dB = 20 * np.log10(magnitude)
phase = np.angle(H)

# CÃ¡lculo da resposta ao impulso
system = lti(b, a)
t, h = impulse(system)

# Plot da resposta em frequÃªncia
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.plot(w / np.pi, dB, 'b')
plt.title('Resposta em FrequÃªncia do Filtro Butterworth')
plt.xlabel('FrequÃªncia Normalizada (x Ï€ rad/sample)')
plt.ylabel('Magnitude (dB)')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(w / np.pi, phase, 'r')
plt.title('Fase da Resposta em FrequÃªncia')
plt.xlabel('FrequÃªncia Normalizada (x Ï€ rad/sample)')
plt.ylabel('Fase (radianos)')
plt.grid()

# Plot da resposta ao impulso
plt.figure(figsize=(8, 6))
plt.stem(t, h, basefmt=" ")
plt.title('Resposta ao Impulso do Filtro Butterworth')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid()

plt.tight_layout()
plt.show()
```

###SIMULAÃ‡ÃƒO 8.2

```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

# ParÃ¢metros do filtro
Wp = 0.2 * np.pi  # FrequÃªncia de passagem (rad/s)
Ws = 0.3 * np.pi  # FrequÃªncia de rejeiÃ§Ã£o (rad/s)
Rp = 7            # Ripple na banda de passagem (dB)
As = 16           # AtenuaÃ§Ã£o na banda de rejeiÃ§Ã£o (dB)

# ConversÃ£o de Ripple e AtenuaÃ§Ã£o para fatores lineares
Ripple = 10 ** (-Rp / 20)
Attn = 10 ** (-As / 20)

# CÃ¡lculo da ordem do filtro e da frequÃªncia de corte
N, Wn = signal.buttord(Wp/np.pi, Ws/np.pi, Rp, As)

# Coeficientes do filtro Butterworth
b, a = signal.butter(N, Wn, btype='low')

# CÃ¡lculo da resposta em frequÃªncia
w, h = signal.freqs(b, a, worN=2000)

# CÃ¡lculo da resposta ao impulso
t, impulse_response = signal.impulse((b, a))

# GrÃ¡fico da resposta em frequÃªncia
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(w * np.pi, 20 * np.log10(abs(h)), 'b')
plt.title('Resposta em FrequÃªncia do Filtro IIR Butterworth')
plt.xlabel('FrequÃªncia (rad/s)')
plt.ylabel('Magnitude (dB)')
plt.grid()

# GrÃ¡fico da resposta ao impulso
plt.subplot(2, 1, 2)
plt.stem(t, impulse_response, basefmt=" ", linefmt="r")
plt.title('Resposta ao Impulso do Filtro IIR Butterworth')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid()

plt.tight_layout()
plt.show()
```
